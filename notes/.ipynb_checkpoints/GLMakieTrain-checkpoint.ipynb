# GLMakie


### Startup

```julia
using GLMakie

f = Figure(backgroudcolor = :tomato, resolution = (600, 300))
ax = Axis(f[1,1],
    title = "Hello, Makie!",
    xlabel = "time(s)",
    ylabel = "quantity(g)"
)
x = range(0, 10, length = 30)
y₁ = 1 ./ log.(x.^2)
y₂ = ℯ .^ (1 ./ (x .+ 1)) .- 3
lines!(ax, x, y₁)
scatter!(ax, x, y₂; color = "#0f0", markersize = 5)
f
```

### Inline

```julia
x₀ = range(0, 10, length=80)
y₃ = sin.(x₀)
figure, axis, lineplot = lines(x₀, y₃,
    figure = (; resolution = (1600, 900)),
    axis = (; title = "Sin Line", xlabel = "x₀ label")
)
lines!(axis, [Point(1,0), Point(2,1), Point(3,0.5)])

ϕ₍x₀₎ = ℯ .^ (.√x₀ ./ (x₀ .+ 5))
dotted = scatter!(
    x₀, ϕ₍x₀₎; 
    marker = :utriangle,
    markersize = range(3, 20, length=length(x₀)),
    color = range(0, 1, length=length(x₀)),
    colormap = :balance
)
dotted.strokewidth = 2
dotted.strokecolor = "#fff"

figure
```

### Layout

```julia
x = LinRange(0, 10, 100)
y = sin.(x)

fig = Figure()
lines(fig[1, 1], x, y, color = :red)
lines(fig[1, 2], x, y, color = :blue)
lines(fig[2, 1:2], x, y, color = :green)

fig
```

### Legend

```julia
fig = Figure()
x₄ = range(0, 10, length=80)
ϕ₍x₄₎ = ℯ .^ (.√x₄ ./ (x₄ .+ 5))
ψ₍x₄₎ = .√x₄
ax1, l1 = lines(fig[1, 1], x₄, ϕ₍x₄₎, color = :red)
ax2, l2 = lines(fig[2, 1], x₄, ψ₍x₄₎, color = :blue)
Legend(fig[1:2, 2], [l1, l2], ["ℯ^(√x₄/(x₄+5))", "√x₄"])
fig
```

## 倍周期分叉

### Logistic模型

$$ 
\begin{align*}
\frac{\mathrm{d}N}{\mathrm{d}t} &= r(1-\frac{N}{K})N 
&= \frac{r(K-N)N}{K} \\
\end{align*}
$$

- 调整公式

$$
\begin{align*}
\frac{K}{(K-N)N}\mathrm{d}N &= r\mathrm{d}t \\
(\frac{1}{K}+\frac{1}{K-N})\mathrm{d}N &= r\mathrm{d}t \\
\ln K+\ln(K-N) &= rt+C_1 \\
K(K-N) &= C_2e^{rt}
\end{align*}
$$

- 差分方程

$$
\begin{align*}
x_{k+1} &= bx_{k}(1-x_{k}) \\
(k &= 0,1,2\dots)
\end{align*}
$$
